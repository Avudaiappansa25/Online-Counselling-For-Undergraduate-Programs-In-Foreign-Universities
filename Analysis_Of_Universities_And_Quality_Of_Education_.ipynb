```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import hypergeom
from scipy.stats import binom
```

```python
#Suppose we have a collection of 20 applications, of which 7 are boys.
# Then if we want to know the probability of finding a given number of boys if we choose at random 12 of the 20 applicantions.
```

```python
[M, n, N] = [20, 7, 12]
rv = hypergeom(M, n, N)
x = np.arange(0, n+1)
pmf_dogs = rv.pmf(x)
fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(x, pmf_dogs, 'bo')
ax.vlines(x, 0, pmf_dogs, lw=2)
ax.set_xlabel('# of boys in our group of chosen applicants')
ax.set_ylabel('hypergeom PMF')
plt.show()
```
